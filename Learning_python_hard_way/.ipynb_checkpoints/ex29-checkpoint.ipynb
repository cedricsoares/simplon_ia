```python
people = 20 
cats = 30 
dogs = 15

if people < cats:
    print("Too many cats! The world is damned!")

if people > cats:
    print("Not many cats! The world is saved!")
    
if people < dogs:
    print("The world is drooled on!")
    
if people > dogs:
    print("The world is dry!")

dogs += 5

if people >= dogs:
    print("People are greater or equal to dogs.")
    
if people <= dogs:
    print("People are less or aqual to dogs.")

if people == dogs:
    print("People are dogs.")
```
