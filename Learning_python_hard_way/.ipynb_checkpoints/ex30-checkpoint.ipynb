```python
people = 30
cars = 40
trucks = 15 

if cars > people: #check if number of car is greatter than people condition is true
    print("We should take the cars.") #then print sentence 
else: #behavior if condition is false
    print("We can't decide.")
    
if trucks > cars: #check if trucks is greatter than cars condition is true
    print("That's too many trucks.") #the print sentence
elif trucks < cars: #if condition is first condition is false elif statement will test the second one : trucks < cars
    print("Maybe we could take the trucks")
else: #if none of previous conditions are true then do
    print("We still can't decide")
    
if people > trucks:
    print("Alright, let's just take the trucks.")
else:
    print("Fine, let's stay home then.")
```
