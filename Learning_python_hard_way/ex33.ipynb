```python
i = 0
numbers = []

while i < 6: 
    print(f"At the top i is {i}")
    numbers.append(i)
    
    i = i + 1
    print("Numbers now: ", numbers)
    print(f"At the bottom i is {i}")
    
print("The numbers: ")

for num in numbers:
    print(num)
```

```python
#Study drills 1 to 4 with a function instead of while loop

numbers = []

def num_list(size, increment, i=0):
    
    print(f"At the top i is: {i}")
    numbers.append(i)
    
    i += increment
    print("Numbers now: ", numbers)
    print(f"At the bottom i is: {i}")
    
    
    if size != i:
        num_list(size, increment, i)
        

num_list(6, 2)

print("The numbers")

for num in numbers:
    print(num)    
```

```python
# Study drill 5 with a range loop

size = 6
numbers = []

for i in range (0, size):
    print(f"At the top i is: {i}")
    numbers.append(i)
    print("Numbers now: ", numbers)
    print(f"at the botom i is: {i+1}")
    
for num in numbers:
    print(num)
    
```
