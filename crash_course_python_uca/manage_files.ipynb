```python
 f = open('./test_file.txt', 'w', encoding='utf8') #Open a file with write rights encoding in utf8
```

```python
for i in range (100):
    f.write(f"ligne{i+1}\n")
```

```python
f.close()
```

```python
f = open('./test_file.txt', 'r', encoding ='utf8')
```

```python
f2 = open('./test_file2.txt', 'w', encoding ='utf8')
```

```python
for line in f : #It is possible to iterate by line in a file
    line = line.split()
    line[0] = line[0].upper()
    f2.write(",".join(line) + "\n")
```

```python
f.close()
```

```python
f2.close()
```

```python
with open('./test_file.txt', 'r', encoding='utf8') as f: #Using a context manager, it is not necessary to close a file
    for line in f :
        print(line)
```

```python
with open(r'./test.bin', 'bw' ) as f: #b parameter to open a file in binary mode
    for i in range(100):
        f.write(b'\x3d') #b parameter to write un binary format
```
