```python
import numpy as np
```

```python
print(np.array([1,2,3])) # declare a normal array filled by a list
```

```python
z = np.zeros(10) # declare 10 size zero array / vector
print(z)
```

```python
z = np.zeros((10,10)) # declare a 10x10 zero matrix
print(z)
```

```python
o = np.ones((10,10)) # declare a 10x10 ones array
print(o)
```

```python
r = np.random.random((10,10)) # declare a 10x10 random matrix
print(r)
```

```python
G = np.random.randn(10,10) # declare a 10x10 random matrix with a normal (gaussian) distribution
print(G)
```

```python
print(G.mean()) # print the mean of a G
```

```python
print(G.var()) # print the variance of a G
```
